# Creating Numpy Array of Available Players

```python
url = 'p:\Mitchell\MockDraft\player_rankings_2025.csv'

raw = pd.read_csv(url, na_values='?', comment='\t', skipinitialspace=True)

playerData = np.array(raw)

dataS = playerData.shape   # array size

def available(availablePlayers):
   dataS = availablePlayers.shape   # array size

   return playerData

def availableRBs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   runningbacks = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "RB"):
         runningbacks.append(availablePlayers[h])

   RBs = np.array(runningbacks)

   return RBs

def availableWRs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   widereceivers = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "WR"):
         widereceivers.append(availablePlayers[h])

   WRs = np.array(widereceivers)

   return WRs

def availableQBs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   quaterbacks = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "QB"):
         quaterbacks.append(availablePlayers[h])

   QBs = np.array(quaterbacks)

   return QBs

def availableTEs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   tightends = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "TE"):
         tightends.append(availablePlayers[h])

   TEs = np.array(tightends)

   return TEs

def availableKs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   kickers = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "K"):
         kickers.append(availablePlayers[h])

   Ks = np.array(kickers)

   return Ks

def availableDSTs(availablePlayers):
   dataS = availablePlayers.shape   # array size

   defense = []
   
   for h in range(dataS[0]):
      player = availablePlayers[h][2]
      if(player == "DST"):
         defense.append(availablePlayers[h])

   DSTs = np.array(defense)

   return DSTs

# print(playerData[0][1])
```

# Display Window

```python
count = 1    
buttons_names = []
textBox = []
teams = {
     "Matt" : [],
     "Mitchell" : [],
     "Caleb" : [],
     "Carter" : [],
     "Trey" : [],
     "Damon" : [],
     "Dalton" : [],
     "Chance" : [],
     "Chris" : [],
     "Jake" : [],
     "Kyle" : [],
     "Zach" : []
}
previous = []

def show_rbs(playersRB, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingRBs = availableRBs(playersRB)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingRBs, playersAll, frameMenu)

def show_wrs(playersWR, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingWRs = availableWRs(playersWR)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingWRs, playersAll, frameMenu)

def show_qbs(playersQB, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingQBs = availableQBs(playersQB)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingQBs, playersAll, frameMenu)

def show_tes(playersTE, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingTEs = availableTEs(playersTE)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingTEs, playersAll, frameMenu)

def show_ks(playersK, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingKs = availableKs(playersK)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingKs, playersAll, frameMenu)

def show_dsts(playersDST, frameButtons, frameOdds, frameMenu, playersAll):
     global buttons_names

     remainingDTSs = availableDSTs(playersDST)
     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButtons, frameOdds, remainingDTSs, playersAll, frameMenu)

def draft(name, size, playersDraft, frameButtons, frameOdds, frameMenu):
     global count
     global teams
     global previous
     round = 0
     pickNum = 0

     for i in range(size):
          playerName = playersDraft[i][1]
          if(name == playerName):
               pos = playersDraft[i][2]
               playersDraft = np.delete(playersDraft, i, axis=0)


     round = count/12.0
     round = math.ceil(round)

     pickNum = abs(((round * 12) - count) - 12)
     user = draft_order_2025[pickNum-1]

     if(len(previous) >= 10):
          previous.pop(0)

     insert = [name, pos]
     previous.append(insert)

     teams[user].append(insert)
     print(insert)
     count += 1
     print("total pick number is", count)
     draft_update(playersDraft, frameButtons, frameOdds, frameMenu)

def odds_update(frameOdds, frameButtons, playersOdds, frameMenu):
     global textBox
     
     for i in range(len(textBox)):
          textBox[i].destroy()

     create_odds(frameOdds, playersOdds, frameButtons, frameMenu)

def draft_update(playersUpdate, frameButtons, frameOdds, frameMenu):
     global buttons_names

     for i in range(len(buttons_names)):
          buttons_names[i].destroy()


     # options = tk.Toplevel()  # Create a new Toplevel window
     # options.title("Drafting")
     # options.geometry("750x750")

     odds_update(frameOdds, frameButtons, playersUpdate, frameMenu)
     # create_odds(frameOdds, playersUpdate, count, frameButtons)

def view_more(PlayersMore, frameButton, frameOdds, playersAll, frameMenu):
     global buttons_names

     for i in range(len(buttons_names)):
          buttons_names[i].destroy()

     create_buttons(frameButton, frameOdds, PlayersMore[37:], playersAll, frameMenu)
     
def menu_creation(frameMenu, frameButtons, playersMenu, frameOdds, playersAll):
     global count

     playerOptions = tk.Menu(frameMenu)
     frameMenu.config(menu=playerOptions)

     player_options = tk.Menu(playerOptions, tearoff=0)

     playerOptions.add_cascade(label="Player Options", menu=player_options)

     player_options.add_command(label="RB Options", command=lambda :show_rbs(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))
     player_options.add_command(label="WR Options", command=lambda :show_wrs(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))
     player_options.add_command(label="QB Options", command=lambda :show_qbs(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))
     player_options.add_command(label="TE Options", command=lambda :show_tes(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))
     player_options.add_command(label="K Options", command=lambda :show_ks(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))
     player_options.add_command(label="DST Options", command=lambda :show_dsts(playersMenu, frameButtons, frameOdds, frameMenu, playersAll))

def create_buttons(frameButtons, frameOdds, playersCreate, playersAll, frameMenu):
     global buttons_names
     global textBox
     global count
     size = dataS[0]

     if(size < 36):
          S = size
     else:
          S = 36

     c = 1
     for i in range(S):
          r = int(i/4)+1
          if(c > 4):
               c = 1
          # Creating a button with specified options
          player_button = tk.Button(frameButtons, 
                    text=playersCreate[i][1], 
                    command=lambda j=playersCreate[i][1]: draft(j, S, playersAll, frameButtons, frameOdds, frameMenu),
                    activebackground="blue", 
                    activeforeground="white",
                    anchor="center",
                    bd=3,
                    bg="lightgray",
                    cursor="hand2",
                    disabledforeground="gray",
                    fg="black",
                    font=("Arial", 12),
                    height=2,
                    highlightbackground="black",
                    highlightcolor="green",
                    highlightthickness=2,
                    justify="center",
                    overrelief="raised",
                    padx=10,
                    pady=5,
                    width=15,
                    wraplength=100
                    )
          player_button.grid(row = r, column= c, padx=5, pady =5)
          buttons_names.append(player_button)
          c += 1

     more_button = tk.Button(frameButtons,
               text="view more players", 
               command=lambda j=playersCreate[i][1]: view_more(playersCreate, frameButtons, frameOdds, playersAll, frameMenu),
               activebackground="blue", 
               activeforeground="white",
               anchor="center",
               bd=3,
               bg="lightgray",
               cursor="hand2",
               disabledforeground="gray",
               fg="black",
               font=("Arial", 12),
               height=2,
               highlightbackground="black",
               highlightcolor="green",
               highlightthickness=2,
               justify="center",
               overrelief="raised",
               padx=10,
               pady=5,
               width=15,
               wraplength=100
               )
     
     more_button.grid(row = r+2, column= 1, padx=5, pady =5)
     menu_creation(frameMenu, frameButtons, playersCreate, frameOdds, playersAll)

def create_odds(frameOdds, playersOdds, frameButtons, frameMenu):
     global count
     global textBox
     global teams
     global previous
     round = 0
     pickNum = 0
     textBox = []

     roundList = ["Round1", "Round2", "Round3", "Round4", "Round5", "Round6", "Round7", "Round8", "Round9", "Round10", "Round11", "Round12", "Round13", "Round14", "Round15", "Round16"]
     oddsList = ["Odds1", "Odds2", "Odds3", "Odds4", "Odds5", "Odds6", "Odds7", "Odds8", "Odds9", "Odds10", "Odds11", "Odds12", "Odds13", "Odds14", "Odds15", "Odds16"]

     round = count/12.0
     round = math.ceil(round)
     print("round", round)

     pickNum = abs(((round * 12) - count) - 12)
     print("pick", pickNum)
     user = picks[pickNum-1]
     teamUser = draft_order_2025[pickNum-1]

     roundNum = roundList[round-1]
     oddsNum = oddsList[round-1]

     nextPlayerName = user[roundNum]
     nextPlayerName = nextPlayerName.dropna(axis = 0)
     nextPlayerOdds = user[oddsNum]
     nextPlayerOdds = nextPlayerOdds.dropna(axis = 0)

     player_label = tk.Label(frameOdds, text = draft_order_2025[pickNum-1] + " Player Odds")

     my_text_block = tk.Text(frameOdds, height=15, width=30)
     my_text_block.insert(tk.END, nextPlayerName + " " + nextPlayerOdds)

     roster_block = tk.Text(frameOdds, height=15, width=25)
     for player in teams[teamUser]:
          for i, item in enumerate(player):
               roster_block.insert(tk.END, item + " ")
          roster_block.insert(tk.END, '\n')

     drafted_block = tk.Text(frameOdds, height=15, width=25)
     for item in previous:
          for p, player in enumerate(item):
               drafted_block.insert(tk.END, player + ' ')
          drafted_block.insert(tk.END, '\n')
     
     player_label.pack()
     my_text_block.pack()
     roster_block.pack()
     drafted_block.pack()

     textBox.append(player_label)
     textBox.append(my_text_block)
     textBox.append(roster_block)
     textBox.append(drafted_block)
   
     create_buttons(frameButtons, frameOdds, playersOdds, playersOdds, frameMenu)

def new_window(players):

     options = tk.Toplevel()  # Create a new Toplevel window
     options.title("Drafting")
     options.geometry("1000x750")

     left_frame = tk.Frame(options, bg="lightblue", bd=2, relief="groove")    
     left_frame.pack(side = 'left', fill='both', expand=True)

     right_frame = tk.Frame(options, bg="lightgreen", bd=2, relief="groove")    
     right_frame.pack(side = 'right', fill='both', expand=True)
 
     close_button = tk.Button(right_frame,
               text="CLOSE", 
               command=lambda : close(),
               activebackground="blue", 
               activeforeground="white",
               anchor="center",
               bd=3,
               bg="lightgray",
               cursor="hand2",
               disabledforeground="gray",
               fg="black",
               font=("Arial", 12),
               height=2,
               highlightbackground="black",
               highlightcolor="green",
               highlightthickness=2,
               justify="center",
               overrelief="raised",
               padx=10,
               pady=5,
               width=15,
               wraplength=100
               )
     
     close_button.pack(side=BOTTOM)

     create_odds(right_frame, players, left_frame, options)

def start():
     global count
     count = 1
     playersStart = available(playerData)
     new_window(playersStart)

def close():
     root.destroy()

root = tk.Tk()

my_frame = tk.Frame(root, bg="grey", bd=2, relief="groove")    
my_frame.pack(pady=10, padx=10, fill='both', expand=True)

root.title("Draft Test")
root.configure(background = "black")
root.minsize(350,350)
# root.maxsize(750,750)
root.geometry("250x250+50+50")

start_button = tk.Button(my_frame, 
     text="Start", 
     command=lambda: start(),
     background="black",
     fg = "white",
     height = 3,
     width = 25,
     )
start_button.place(relx=0.5, rely=0.5,anchor=tk.CENTER)   

root.mainloop()
```
