```python
url = 'p:\Mitchell\MockDraft\player_rankings_2025.csv'

raw = pd.read_csv(url, na_values='?', comment='\t', skipinitialspace=True)

playerData = np.array(raw)

dataS = playerData.shape   # array size

def available():
   dataS = playerData.shape   # array size

   runningbacks = []
   widereceivers = []
   quaterbacks = []
   tightends = []
   kickers = []
   defense = []

   for h in range(dataS[0]):
      player = playerData[h][2]
      if(player == "RB"):
         runningbacks.append(playerData[h])
      if(player == "WR"):
         widereceivers.append(playerData[h])
      if(player == "QB"):
         quaterbacks.append(playerData[h])
      if(player == "TE"):
         tightends.append(playerData[h])
      if(player == "K"):
         kickers.append(playerData[h])
      if(player == "D/ST"):
         defense.append(playerData[h])

   RBs = np.array(runningbacks)
   WRs = np.array(widereceivers)
   QBs = np.array(quaterbacks)
   TEs = np.array(tightends)
   Ks = np.array(kickers)
   DST = np.array(defense)

   return playerData


# print(playerData[0][1])
```

```python
global count

def next_odds(frameOdds, playersOdds, counter, frameButtons):
     round = 0
     pickNum = 0
     textBox = []

     roundList = ["Round1", "Round2", "Round3", "Round4", "Round5", "Round6", "Round7", "Round8", "Round9", "Round10", "Round11", "Round12", "Round13", "Round14", "Round15", "Round16"]
     oddsList = ["Odds1", "Odds2", "Odds3", "Odds4", "Odds5", "Odds6", "Odds7", "Odds8", "Odds9", "Odds10", "Odds11", "Odds12", "Odds13", "Odds14", "Odds15", "Odds16"]

     round = counter/12.0
     round = math.ceil(round)
     print("round", round)

     pickNum = abs(((round * 12) - counter) - 12)
     print("pick", pickNum)
     user = picks[pickNum-1]

     roundNum = roundList[round-1]
     oddsNum = oddsList[round-1]

     nextPlayerName = user[roundNum]
     nextPlayerName = nextPlayerName.dropna(axis = 0)
     nextPlayerOdds = user[oddsNum]
     nextPlayerOdds = nextPlayerOdds.dropna(axis = 0)

     player_label = tk.Label(frameOdds, text = draft_order_2025[pickNum-1] + " Player Odds")

     my_text_block = tk.Text(frameOdds, height=10, width=50)
     my_text_block.insert(tk.END, nextPlayerName + " " + nextPlayerOdds)
     
     player_label.pack()
     my_text_block.pack()

     textBox.append(player_label)
     textBox.append(my_text_block)

     create_buttons(frameButtons, frameOdds, playersOdds, counter, textBox, playersOdds)

def create_buttons(frameButtons, frameOdds, playersCreate, counter, textBox, playersAll):
     buttons_names = []
     size = dataS[0]
     count = counter

     if(size < 36):
          S = size
     else:
          S = 36

     c = 1
     for i in range(S):
          r = int(i/4)+1
          if(c > 4):
               c = 1
          # Creating a button with specified options
          player_button = tk.Button(frameButtons, 
                    text=playersCreate[i][1], 
                    command=lambda j=playersCreate[i][1]: draft(j, S, playersAll, buttons_names, frameButtons, count, frameOdds, textBox),
                    activebackground="blue", 
                    activeforeground="white",
                    anchor="center",
                    bd=3,
                    bg="lightgray",
                    cursor="hand2",
                    disabledforeground="gray",
                    fg="black",
                    font=("Arial", 12),
                    height=2,
                    highlightbackground="black",
                    highlightcolor="green",
                    highlightthickness=2,
                    justify="center",
                    overrelief="raised",
                    padx=10,
                    pady=5,
                    width=15,
                    wraplength=100
                    )
          player_button.grid(row = r, column= c, padx=5, pady =5)
          buttons_names.append(player_button)
          c += 1

     more_button = tk.Button(frameButtons,
               text="view more players", 
               command=lambda j=playersCreate[i][1]: view_more(playersCreate, buttons_names, frameButtons, frameOdds, count, textBox, playersAll),
               activebackground="blue", 
               activeforeground="white",
               anchor="center",
               bd=3,
               bg="lightgray",
               cursor="hand2",
               disabledforeground="gray",
               fg="black",
               font=("Arial", 12),
               height=2,
               highlightbackground="black",
               highlightcolor="green",
               highlightthickness=2,
               justify="center",
               overrelief="raised",
               padx=10,
               pady=5,
               width=15,
               wraplength=100
               )
     
     more_button.grid(row = r+2, column= 1, padx=5, pady =5)

def draft(name, size, playersDraft, button, frameButtons, counter, frameOdds, textBox):
     count = counter

     for i in range(size):
          playerName = playersDraft[i][1]
          if(name == playerName):
               playersDraft = np.delete(playersDraft, i, axis=0)
     # print(playersDraft)
     print(name)
     count += 1
     print("total pick number is", count)
     draft_update(playersDraft, button, frameButtons, count, frameOdds, textBox)

def odds_update(frameOdds, frameButtons, playersOdds, counter, textBox):
     
     for i in range(len(textBox)):
          textBox[i].destroy()

     next_odds(frameOdds, playersOdds, counter, frameButtons)

def draft_update(playersUpdate, button, frameButtons, counter, frameOdds, textBox):
     count = counter
     # old_window.destroy()  # Close the old window
     for i in range(len(button)):
          button[i].destroy()


     # options = tk.Toplevel()  # Create a new Toplevel window
     # options.title("Drafting")
     # options.geometry("750x750")

     odds_update(frameOdds, frameButtons, playersUpdate, counter, textBox)
     # next_odds(frameOdds, playersUpdate, count, frameButtons)

def view_more(PlayersMore, button, frameButton, frameOdds, counter, textBox, playersAll):
     # old_window.destroy()  # Close the old window
     for i in range(len(button)):
          button[i].destroy()

     create_buttons(frameButton, frameOdds, PlayersMore[37:], counter, textBox, playersAll)

def new_window(players, counter):

     options = tk.Toplevel()  # Create a new Toplevel window
     options.title("Drafting")
     options.geometry("1000x750")

     left_frame = tk.Frame(options, bg="lightblue", bd=2, relief="groove")    
     left_frame.pack(side = 'left', fill='both', expand=True)

     right_frame = tk.Frame(options, bg="lightgreen", bd=2, relief="groove")    
     right_frame.pack(side = 'right', fill='both', expand=True)

     next_odds(right_frame, players, counter, left_frame)

def start():
     count = 1
     playersStart = available()
     new_window(playersStart, count)

root = tk.Tk()

my_frame = tk.Frame(root, bg="grey", bd=2, relief="groove")    
my_frame.pack(pady=10, padx=10, fill='both', expand=True)

root.title("Draft Test")
root.configure(background = "black")
root.minsize(350,350)
# root.maxsize(750,750)
root.geometry("250x250+50+50")

start_button = tk.Button(my_frame, 
     text="Start", 
     command=lambda: start(),
     background="black",
     fg = "white",
     height = 3,
     width = 25,
     )
start_button.place(relx=0.5, rely=0.5,anchor=tk.CENTER)   


root.mainloop()
```

```python
import tkinter as tk

def create_new_button(parent_frame, button_count):
    """Creates and places a new button."""
    new_button = tk.Button(parent_frame, text=f"Button {button_count}")
    new_button.pack(pady=5) # Example placement

def add_button_handler():
    """Handler for the initial button click."""
    global button_counter # Assuming button_counter is a global variable
    button_counter += 1
    create_new_button(root, button_counter)

root = tk.Tk()
root.title("Dynamic Button Creation")

button_counter = 0 # Initialize a counter for new buttons

# Initial button to trigger creation
initial_button = tk.Button(root, text="Create New Button", command=add_button_handler)
initial_button.pack(pady=10)

root.mainloop()
```
